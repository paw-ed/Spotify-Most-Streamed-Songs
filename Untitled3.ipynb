```python
import plotly.express as px

# Assuming you have a DataFrame called 'data_df' and you want to remove outliers from it
# and store it in 'df_no_outliers'

# Example outlier removal using IQR (Interquartile Range) method
Q1 = data_df['danceability_%'].quantile(0.25)
Q3 = data_df['danceability_%'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = data_df[(data_df['danceability_%'] >= lower_bound) & (data_df['danceability_%'] <= upper_bound)]


# Scatter plot for danceability vs streams
fig2 = px.scatter(df_no_outliers,
                  x='danceability_%',
                  y='streams',
                  color='artist_count',
                  hover_data=['track_name', 'artist(s)_name'],
                  title='Danceability vs Streams',
                  color_continuous_scale='Plasma')
fig2.show()
```

```python
fig3 = px.histogram(df_no_outliers,
                    x='bpm',
                    title='Distribution of BPM',
                    color='released_year',
                    color_discrete_sequence=px.colors.qualitative.Plotly)
fig3.show()
```

```python
correlation_matrix = df_no_outliers.select_dtypes(include=['float64', 'int64']).corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
fig = px.scatter(df_no_outliers,
                  x='danceability_%',
                  y='energy_%',
                  color='released_year',
                  hover_data=['track_name', 'artist(s)_name'],
                  title='Energy Levels vs Danceability',
                  color_continuous_scale='YlGnBu')
fig.show()
```

```python
import plotly.express as px

fig = px.scatter_matrix(df_no_outliers,
                          dimensions=['danceability_%', 'energy_%', 'valence_%', 'acousticness_%', 'bpm'],
                          title='Pair Plot of Selected Features',
                          color='released_year',
                          color_continuous_scale='Cividis')
fig.show()
```

```python
release_trends = df_no_outliers['released_year'].value_counts().sort_index().reset_index()
release_trends.columns = ['released_year', 'song_count']

fig = px.line(release_trends,
                x='released_year',
                y='song_count',
                title='Trends in Song Releases Over the Years',
                markers=True,
                color='song_count')
fig.show()
```

```python
fig = px.scatter(df_no_outliers,
                  x='liveness_%',
                  y='streams',
                  color='released_year',
                  hover_data=['track_name', 'artist(s)_name'],
                  title='Liveness vs Streams',
                  color_continuous_scale='YlOrRd')
fig.show()
```

```python
fig = px.scatter(df_no_outliers,
                  x='acousticness_%',
                  y='instrumentalness_%',
                  color='released_year',
                  hover_data=['track_name', 'artist(s)_name'],
                  title='Acousticness vs Instrumentalness',
                  color_continuous_scale='Sunset')
fig.show()
```

```python
key_counts = df_no_outliers['key'].value_counts().reset_index()
key_counts.columns = ['Key', 'Count']

fig = px.bar(key_counts,
               x='Key',
               y='Count',
               title='Frequency of Each Musical Key',
               color='Count',
               color_continuous_scale='Viridis')
fig.show()
```

```python
# Scatter plot for speechiness vs streams
fig = px.scatter(df_no_outliers,
                  x='speechiness_%',
                  y='streams',
                  color='released_year',
                  hover_data=['track_name', 'artist(s)_name'],
                  title='Speechiness vs Streams',
                  color_continuous_scale='Icefire')
fig.show()
```

```python
fig = px.scatter(df_no_outliers,
                  x='valence_%',
                  y='energy_%',
                  color='released_year',
                  hover_data=['track_name', 'artist(s)_name'],
                  title='Valence vs Energy Levels',
                  color_continuous_scale='YlGn')
fig.show()
```

```python
# Scatter plot for acousticness vs danceability
fig = px.scatter(df_no_outliers,
                  x='acousticness_%',
                  y='danceability_%',
                  color='released_year',
                  hover_data=['track_name', 'artist(s)_name'],
                  title='Acousticness vs Danceability',
                  color_continuous_scale='Reds')
fig.show()
```

```python
# Line chart showing trends in artist count by released year
artist_trends = df_no_outliers.groupby('released_year')['artist_count'].mean().reset_index()

fig = px.line(artist_trends,
                x='released_year',
                y='artist_count',
                title='Trends in Number of Artists Contributing to Songs Over the Years',
                markers=True,
                color='artist_count')
fig.show()
```

```python
# Violin plot for acousticness by presence in charts
fig = px.violin(df_no_outliers,
                 x='in_spotify_charts',
                 y='acousticness_%',
                 title='Distribution of Acousticness by Presence in Spotify Charts',
                 color='in_spotify_charts',
                 box=True)
fig.show()
```
